```python
def get_futures_order_book(client, pair, upto=None, pct=None):
    """
    upto returns the sum between current price up to upto price
    pct returns the sum between current price and pct*current price
    """
    book = client.futures_order_book(symbol=pair, limit=1000)
    asks = list(map(lambda x: [float(x[0]), float(x[1])], book['asks']))
    price = asks[0][0]*pct if upto is None else upto
    asks = list(filter(lambda x: x[0]<price, asks))
    return sum(list(map(lambda x: x[1], asks)))
```

```python
get_futures_order_book(client, "XTZUSDT", pct=1.001)
```
